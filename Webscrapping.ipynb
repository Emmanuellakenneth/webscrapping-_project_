## scrape data for multiple pages

```python
for page in range(1,201): # use a for loop to get multiple pages
    res=requests.get(f'https://www.cars45.com/listing?page={page}')
    print (res.url)
```

```python
# EMPTY DICTIONARY

car_data={
    'car_id':[],
    'amount':[],
    'name':[],
    'region':[],
    'use_type':[],
    'mileage':[],
}
```

```python
for page in range(1,201):
    res=requests.get(f'https://www.cars45.com/listing?page={page}')
    
    #create an object of the beautifulsoup
    soup=BeautifulSoup(res.content,"html.parser")
    
    car15=soup.find_all('a',class_='car-feature car-feature--wide-mobile')
    
    for car in car15:
        car_id= car['href']
        amount=car.find('p',class_='car-feature__amount').get_text().strip()
        name=car.find('p',class_='car-feature__name').get_text().strip()
        region=car.find('p',class_='car-feature__region').get_text().strip()
        use_type=car.find('span',class_='car-feature__others__item').get_text().strip()
        mileage=use_type=car.find('span',class_='car-feature__others__item').find_next('span').get_text().strip()

    #store data in a car_data
    car_data['car_id'].append(car_id)
    car_data['amount'].append(amount)
    car_data['name'].append(name)
    car_data['region'].append(region)
    car_data['use_type'].append(use_type)
    car_data['mileage'].append(mileage)
```

```python
car_df=pd.DataFrame(car_data)
```
